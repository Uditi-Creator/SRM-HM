```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from scipy import stats

# Setting style for all plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# creating a category column from the one-hot encoded columns
category_columns = ['category_Cabinets', 'category_Chairs', 'category_Sofas', 'category_Tables']
df['category'] = df[category_columns].idxmax(axis=1).str.replace('category_', '')

# Creating EDA visualizations for each category
categories = ['Cabinets', 'Chairs', 'Sofas', 'Tables']

for category in categories:
    print(f"\n=== Creating EDA Visualizations for {category} ===")
    
    # Filtering data for this category
    category_data = df[df['category'] == category].copy()
    
    if len(category_data) == 0:
        print(f"No data found for {category}")
        continue
    
    # Creating a figure with multiple subplots
    fig = plt.figure(figsize=(20, 15))
    fig.suptitle(f'Exploratory Data Analysis for {category}', fontsize=16, fontweight='bold')
    
    # 1. Time Series Plot of Sales
    ax1 = plt.subplot(3, 3, 1)
    if 'date' in category_data.columns:
        category_data = category_data.sort_values('date')
        plt.plot(category_data['date'], category_data['sales_units'])
        plt.title('Sales Units Over Time')
        plt.xlabel('Date')
        plt.ylabel('Sales Units')
        plt.xticks(rotation=45)
    else:
        plt.plot(category_data.index, category_data['sales_units'])
        plt.title('Sales Units')
        plt.xlabel('Index')
        plt.ylabel('Sales Units')
    
    # 2. Distribution of Sales Units
    ax2 = plt.subplot(3, 3, 2)
    sns.histplot(category_data['sales_units'], kde=True)
    plt.title('Distribution of Sales Units')
    plt.xlabel('Sales Units')
    
    # 3. Boxplot of Sales by Day of Week
    ax3 = plt.subplot(3, 3, 3)
    if 'day_of_week' in category_data.columns:
        sns.boxplot(x='day_of_week', y='sales_units', data=category_data)
        plt.title('Sales by Day of Week')
        plt.xlabel('Day of Week')
        plt.ylabel('Sales Units')
    
    # 4. Price vs Sales Scatter Plot
    ax4 = plt.subplot(3, 3, 4)
    if 'price' in category_data.columns:
        plt.scatter(category_data['price'], category_data['sales_units'], alpha=0.6)
        plt.title('Price vs Sales Units')
        plt.xlabel('Price')
        plt.ylabel('Sales Units')
        
        # Add trend line
        z = np.polyfit(category_data['price'], category_data['sales_units'], 1)
        p = np.poly1d(z)
        plt.plot(category_data['price'], p(category_data['price']), "r--")
    
    # 5. Discount Impact on Sales
    ax5 = plt.subplot(3, 3, 5)
    if 'discount_percentage' in category_data.columns:
        sns.boxplot(x=pd.cut(category_data['discount_percentage'], bins=5), y='sales_units', data=category_data)
        plt.title('Sales by Discount Percentage')
        plt.xlabel('Discount Percentage Bins')
        plt.ylabel('Sales Units')
        plt.xticks(rotation=45)
    
    # 6. Economic Index vs Sales
    ax6 = plt.subplot(3, 3, 6)
    if 'economic_index' in category_data.columns:
        plt.scatter(category_data['economic_index'], category_data['sales_units'], alpha=0.6)
        plt.title('Economic Index vs Sales Units')
        plt.xlabel('Economic Index')
        plt.ylabel('Sales Units')
        
        # Add trend line
        z = np.polyfit(category_data['economic_index'], category_data['sales_units'], 1)
        p = np.poly1d(z)
        plt.plot(category_data['economic_index'], p(category_data['economic_index']), "r--")
    
    # 7. Seasonal Decomposition (if we have date information)
    ax7 = plt.subplot(3, 3, 7)
    if 'date' in category_data.columns and len(category_data) > 50:
        try:
            # Ensuring the date column is in datetime format
            category_data['date'] = pd.to_datetime(category_data['date'])
            # Setting date as index
            ts_data = category_data.set_index('date')['sales_units']
            # Resampling to daily frequency if needed
            ts_data = ts_data.resample('D').mean().ffill()
            
            # Decomposing time series
            decomposition = seasonal_decompose(ts_data, period=30)  # Assuming monthly seasonality
            
            plt.plot(decomposition.trend)
            plt.title('Trend Component of Sales')
            plt.xlabel('Date')
            plt.ylabel('Trend')
        except:
            plt.text(0.5, 0.5, 'Insufficient data for decomposition', 
                    horizontalalignment='center', verticalalignment='center')
            plt.title('Trend Component (Not Available)')
    
    # 8. Correlation Heatmap for Top Features
    ax8 = plt.subplot(3, 3, 8)
    # Selecting numerical columns for correlation
    numerical_cols = category_data.select_dtypes(include=[np.number]).columns
    # Removing columns with low variance or too many missing values
    numerical_cols = [col for col in numerical_cols if category_data[col].nunique() > 1 and 
                     category_data[col].isnull().mean() < 0.5]
    
    # Calculate correlation matrix
    corr_matrix = category_data[numerical_cols].corr()
    
    # Focus on correlations with sales_units
    sales_corr = corr_matrix['sales_units'].sort_values(ascending=False)
    top_corr_features = sales_corr.index[1:11]  # Top 10 features correlated with sales
    
    # Plot heatmap
    sns.heatmap(category_data[top_corr_features].corr(), annot=True, cmap='coolwarm', center=0)
    plt.title('Top Features Correlation Heatmap')
    plt.xticks(rotation=45)
    
    # 9. Promotion Impact
    ax9 = plt.subplot(3, 3, 9)
    if 'promotion_applied' in category_data.columns:
        sns.boxplot(x='promotion_applied', y='sales_units', data=category_data)
        plt.title('Sales with vs Without Promotion')
        plt.xlabel('Promotion Applied')
        plt.ylabel('Sales Units')
    
    plt.tight_layout()
    plt.savefig(f'{category}_EDA.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Print some statistics
    print(f"\nStatistics for {category}:")
    print(f"Number of records: {len(category_data)}")
    print(f"Average sales: {category_data['sales_units'].mean():.2f}")
    print(f"Sales standard deviation: {category_data['sales_units'].std():.2f}")
    if 'price' in category_data.columns:
        print(f"Average price: {category_data['price'].mean():.2f}")
    if 'discount_percentage' in category_data.columns:
        print(f"Average discount: {category_data['discount_percentage'].mean():.2f}%")

# Additional cross-category comparisons
print("\n=== Cross-Category Comparisons ===")

# 1. Sales distribution across categories
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='sales_units', data=df)
plt.title('Sales Distribution Across Categories')
plt.xlabel('Category')
plt.ylabel('Sales Units')
plt.savefig('Cross_Category_Sales.png', dpi=300, bbox_inches='tight')
plt.show()

# 2. Average price by category
if 'price' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.barplot(x='category', y='price', data=df, estimator=np.mean)
    plt.title('Average Price by Category')
    plt.xlabel('Category')
    plt.ylabel('Average Price')
    plt.savefig('Category_Price_Comparison.png', dpi=300, bbox_inches='tight')
    plt.show()

# 3. Price elasticity across categories
if 'price' in df.columns and 'sales_units' in df.columns:
    plt.figure(figsize=(12, 6))
    for category in categories:
        cat_data = df[df['category'] == category]
        if len(cat_data) > 10:
            # Calculate correlation between price and sales
            correlation = cat_data['price'].corr(cat_data['sales_units'])
            plt.bar(category, correlation, label=f'{category} (r={correlation:.2f})')
    
    plt.title('Price Elasticity by Category (Correlation)')
    plt.xlabel('Category')
    plt.ylabel('Correlation (Price vs Sales)')
    plt.legend()
    plt.savefig('Price_Elasticity_by_Category.png', dpi=300, bbox_inches='tight')
    plt.show()

print("EDA visualizations completed. Check the generated PNG files.")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# Define the function to calculate MAPE safely
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / np.maximum(np.ones(len(y_true)), np.abs(y_true)))) * 100

# Create a category column from the one-hot encoded columns
category_columns = ['category_Cabinets', 'category_Chairs', 'category_Sofas', 'category_Tables']
df['category'] = df[category_columns].idxmax(axis=1).str.replace('category_', '')

# Define features to exclude (potential data leakage + identifiers + target + non-numeric features)
features_to_exclude = [
    'future_demand',           # Direct source of leakage
    'sales_units',             # Target variable
    'date',                    # Identifier
    'product_id',              # Identifier
    'category'                 # The category column we created (contains strings)
] + category_columns           # One-hot encoded category columns

# Add any other features that might be derived from the target
# For example, if you have features like 'next_day_sales' or 'demand_tomorrow'
potential_leakage_features = [col for col in df.columns if any(x in col.lower() for x in ['future', 'next', 'tomorrow', 'ahead'])]
features_to_exclude.extend(potential_leakage_features)

# Also exclude any non-numeric columns that might cause issues
non_numeric_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
features_to_exclude.extend(non_numeric_cols)

print("Features being excluded due to potential data leakage or non-numeric types:")
print(features_to_exclude)

# Define the correct feature set (only numeric columns that are not excluded)
correct_features = [col for col in df.columns if col not in features_to_exclude and df[col].dtype in [np.int64, np.float64]]
print(f"\nNumber of features to use: {len(correct_features)}")
print("Features:", correct_features)

# Dictionary to store results for each category
category_results = {}
category_models = {}  # To store trained models for later use

# List of your categories
categories = ['Cabinets', 'Chairs', 'Sofas', 'Tables']

for category in categories:
    print(f"\n=== Training model for {category} ===")
    
    # Filter data for this category only
    category_data = df[df['category'] == category].copy()
    
    if len(category_data) == 0:
        print(f"No data found for {category}")
        continue
    
    # Prepare features and target
    X = category_data[correct_features]
    y = category_data['sales_units']
    
    # Check if we have enough data
    if len(X) < 10:
        print(f"Not enough data for {category} (only {len(X)} samples)")
        continue
    
    # Split into train/test (use time-based split if it's time series data)
    train_size = int(0.8 * len(X))
    X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
    y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]
    
    # Train model
    model = GradientBoostingRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mape = mean_absolute_percentage_error(y_test, y_pred)
    
    # Store results and model
    category_results[category] = {
        'mae': mae,
        'rmse': rmse,
        'mape': mape,
        'model': model,
        'test_size': len(X_test),
        'train_size': len(X_train)
    }
    category_models[category] = model
    
    print(f"{category} - MAE: {mae:.2f}, RMSE: {rmse:.2f}, MAPE: {mape:.2f}%")
    print(f"Training samples: {len(X_train)}, Test samples: {len(X_test)}")
    
    # Plot feature importance for this category
    plt.figure(figsize=(10, 6))
    feature_importance = pd.DataFrame({
        'feature': X.columns,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    plt.barh(feature_importance['feature'][:10], feature_importance['importance'][:10])
    plt.title(f'Top 10 Feature Importances for {category} (No Data Leakage)')
    plt.xlabel('Importance')
    plt.tight_layout()
    plt.show()

# Print summary results
print("\n=== SUMMARY RESULTS (Without Data Leakage) ===")
for category, results in category_results.items():
    print(f"{category}: MAE={results['mae']:.2f}, MAPE={results['mape']:.2f}%, "
          f"Train samples={results['train_size']}, Test samples={results['test_size']}")

# Compare with previous results (if available)
print("\n=== COMPARISON WITH PREVIOUS RESULTS ===")
print("Note: Previous results had data leakage from 'future_demand'")
print("Category | Previous MAPE | New MAPE | Change")
print("---------|---------------|----------|-------")
for category in categories:
    if category in category_results:
        # These are the previous results you shared
        previous_mape = {
            'Cabinets': 8.81,
            'Chairs': 8.04,
            'Sofas': 8.37,
            'Tables': 9.83
        }.get(category, 'N/A')
        
        new_mape = category_results[category]['mape']
        change = new_mape - previous_mape
        
        print(f"{category:8} | {previous_mape:12.2f}% | {new_mape:7.2f}% | {change:+.2f}%")

# Additional analysis: Check if the models are still useful
print("\n=== MODEL USEFULNESS ASSESSMENT ===")
print("A model is generally considered useful if it performs better than simple benchmarks:")
print("1. Better than predicting the mean (naive baseline)")
print("2. Better than predicting the last value (persistence model)")

for category in categories:
    if category in category_results:
        # Calculate mean baseline
        mean_baseline = np.full_like(y_test, y_train.mean())
        mean_mape = mean_absolute_percentage_error(y_test, mean_baseline)
        
        # Calculate persistence baseline (using the last known value)
        # For time series, this would typically be the previous time period's value
        if 'sales_units_lag_1' in correct_features and len(X_test) > 0:
            persistence_pred = X_test['sales_units_lag_1'].values
        else:
            # Fallback: use the last training value for all test predictions
            persistence_pred = np.full_like(y_test, y_train.iloc[-1])
        
        persistence_mape = mean_absolute_percentage_error(y_test, persistence_pred)
        
        model_mape = category_results[category]['mape']
        
        print(f"\n{category}:")
        print(f"  Model MAPE: {model_mape:.2f}%")
        print(f"  Mean baseline MAPE: {mean_mape:.2f}%")
        print(f"  Persistence baseline MAPE: {persistence_mape:.2f}%")
        
        if model_mape < mean_mape and model_mape < persistence_mape:
            print(f"  ✅ Model is useful (better than baselines)")
        else:
            print(f"  ⚠️  Model may not be useful (worse than baselines)")
```
